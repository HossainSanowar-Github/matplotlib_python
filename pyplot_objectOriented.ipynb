# Figures and Subplots
* up to 2x2=up to 4 plots in total
* up to 3x3=up to 9 plots in total
* selecting first 4 subplots where numbering from 1

```python
fig=plt.figure(figsize=(9,8))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.5)
ax2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
plt.plot(np.random.randn(50).cumsum(),'k--')

plt.show()
```

# Subplots
* create a new figure and returns a Numpy array

```python
fig,axes=plt.subplots(2,2)
```

```python
fig,axes=plt.subplots(2,2,figsize=(8,8),sharex=True,sharey=True)
```

```python
fig,axes=plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(50),bins=20,color='r',alpha=0.5)
plt.subplots_adjust(left=None,bottom=None,right=None, top=None, wspace=0.1,hspace=0.1)
```

# Colors, Markers, and Line Styles

```python
#ax.plot(x,y,'g--')
#ax.plot(x,y,linestyle='--',color='g')

from numpy.random import randn
plt.plot(randn(30).cumsum(),'ko--')

plt.show()
```

```python
data=randn(50).cumsum()
plt.figure(figsize=(10,8))
plt.plot(data,'bo--',label='Default')
plt.plot(data,'k--',drawstyle='steps-post',label='steps-post')
plt.legend(loc='best')
plt.show()
```
